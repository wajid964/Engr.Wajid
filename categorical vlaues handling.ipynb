```python
import pandas as pd
```

```python
df=pd.read_csv('Data.csv')
df
```

```python
df=pd.get_dummies(data=df,columns=['Country'])
df
```

```python
df['Age'].isna().sum()
```

```python
df['Salary'].isna().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Salary'].fillna(df['Salary'].mean(),inplace=True)
```

```python
X=df[['Age','Salary','Country_France','Country_Germany','Country_Spain']]
df['Purchased'] = df['Purchased'].apply(lambda x: 0 if x=='No' else 1)
X
```

```python
y=df[['Purchased']]
y
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
xtr,xte,ytr,yte=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
```

```python
knn.fit(xtr,ytr)
```

```python
y_pred=knn.predict(xte)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,yte))
```
